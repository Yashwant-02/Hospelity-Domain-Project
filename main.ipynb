```python
import os
from six.moves import urllib
import pandas as pd
```

```python
data_dir='./data/'

data_url_list = [r"https://raw.githubusercontent.com/mohitmahi004/atliq_hotels_data/refs/heads/main/fact_bookings.csv",
     r"https://raw.githubusercontent.com/mohitmahi004/atliq_hotels_data/refs/heads/main/dim_date.csv",
     r"https://raw.githubusercontent.com/mohitmahi004/atliq_hotels_data/refs/heads/main/dim_hotels.csv",
     r"https://raw.githubusercontent.com/mohitmahi004/atliq_hotels_data/refs/heads/main/dim_rooms.csv",
     r"https://raw.githubusercontent.com/mohitmahi004/atliq_hotels_data/refs/heads/main/fact_aggregated_bookings.csv"]

for download_url in data_url_list:
    os.makedirs(data_dir, exist_ok=True)
    file_name = os.path.basename(download_url)
    print(file_name)

    download_file_path = os.path.join(data_dir, file_name)
    print(download_file_path)

    urllib.request.urlretrieve(download_url, download_file_path)
```
